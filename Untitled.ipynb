```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
import cv2 as cv
import sys
from scipy.optimize import curve_fit
```

```python
img = cv.imread("crop.png")

cv.imshow("Display window", img)

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)



height, width = gray.shape
cropped_img = gray[0:int(height/2), 0:width]
cropped_img2 = gray[int(height/2):height, 0:width]

cv.imwrite("gray.png",gray)

thresh = 130
im_bw = cv.threshold(cropped_img, thresh, 255, cv.THRESH_BINARY)[1]
cv.imshow("Display window3", im_bw)
thresh2 = 85
im_bw2 = cv.threshold(cropped_img2, thresh2, 255, cv.THRESH_BINARY)[1]
cv.imshow("Display window4", im_bw2)

cv.imwrite("im_bw.png",im_bw)
cv.imwrite("im_bw2.png",im_bw2)

gray[0:int(height/2), 0:width] = im_bw
gray[int(height/2):height, 0:width] = im_bw2
cv.imshow("Display window2", gray)

kernel = np.ones((4, 3), np.uint8)

erosion_dst = cv.erode(gray, kernel)
cv.imshow("Display erosion", erosion_dst)

cv.imwrite("erosion_dst.png",erosion_dst)

erosion_dst = cv.flip(erosion_dst, 0)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
indices = np.where(erosion_dst == [255])
print(indices)
x = indices[1]
y = indices[0]
x_line = np.arange(min(x), max(x), 1)
```

```python
from numpy.polynomial import Chebyshev as T
p = T.fit(x, y, 70)
img = plt.imread("crop.png")
fig, ax = plt.subplots(figsize=(15,10))
ax.imshow(img, extent=[0, width, 0, height])
ax.plot(x, y, '.', x_line, p(x_line), '-')
plt.savefig("better.png",transparent=True)
print(p.convert().coef)
```
